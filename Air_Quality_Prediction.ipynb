```python
data=df
```

```python
#data preprocessing
data = data.drop('Date', axis=1)
data = data.drop('Time', axis=1)
data = data.drop('C6H6(GT)', axis=1)
```

```python
data.head()
```

```python
data.describe()
```

```python
data["CO(GT)"].min()
```

```python
for i in range(0,data.shape[0]):
  if data["CO(GT)"][i]==data["CO(GT)"].min():
     data.loc[i, 'CO(GT)'] = 0
```

```python
data["CO(GT)"].describe()
```

```python
#MAKING THE MINIMUM value to be zero
for i in range(0,data.shape[0]):
  if data["PT08.S1(CO)"][i]==data["PT08.S1(CO)"].min():
     data.loc[i, 'PT08.S1(CO)'] = 0
for i in range(0,data.shape[0]):
  if data["NMHC(GT)"][i]==data["NMHC(GT)"].min():
     data.loc[i, 'NMHC(GT)'] = 0
for i in range(0,data.shape[0]):
  if data["PT08.S2(NMHC)"][i]==data["PT08.S2(NMHC)"].min():
     data.loc[i, 'PT08.S2(NMHC)'] = 0     
for i in range(0,data.shape[0]):
  if data["NOx(GT)"][i]==data["NOx(GT)"].min():
     data.loc[i, 'NOx(GT)'] = 0
for i in range(0,data.shape[0]):
  if data["PT08.S3(NOx)"][i]==data["PT08.S3(NOx)"].min():
     data.loc[i, 'PT08.S3(NOx)'] = 0
for i in range(0,data.shape[0]):
  if data["NO2(GT)"][i]==data["NO2(GT)"].min():
     data.loc[i, 'NO2(GT)'] = 0
for i in range(0,data.shape[0]):
  if data["PT08.S4(NO2)"][i]==data["PT08.S4(NO2)"].min():
     data.loc[i, 'PT08.S4(NO2)'] = 0

for i in range(0,data.shape[0]):
  if data["AH"][i]==data["AH"].min():
     data.loc[i, 'AH'] = 0
for i in range(0,data.shape[0]):
  if data["RH"][i]==data["RH"].min():
     data.loc[i, 'RH'] = 0       
```

```python
for i in range(0,data.shape[0]):
  if data["PT08.S5(O3)"][i]==data["PT08.S5(O3)"].min():
     data.loc[i, 'PT08.S5(O3)'] = 0
```

```python
for i in range(0,data.shape[0]):
  if data["T"][i]==data["T"].min():
     data.loc[i,'T'] = 0
```

```python
data.describe()
```

```python
#filling the minimum values with mean
for i in range(0,data.shape[0]):
  if data["PT08.S1(CO)"][i]==data["PT08.S1(CO)"].min():
     data.loc[i, 'PT08.S1(CO)'] = data["PT08.S1(CO)"].mean()
for i in range(0,data.shape[0]):
  if data["NMHC(GT)"][i]==data["NMHC(GT)"].min():
     data.loc[i, 'NMHC(GT)'] = data["NMHC(GT)"].mean()
for i in range(0,data.shape[0]):
  if data["PT08.S2(NMHC)"][i]==data["PT08.S2(NMHC)"].min():
     data.loc[i, 'PT08.S2(NMHC)'] = data["PT08.S2(NMHC)"].mean()
for i in range(0,data.shape[0]):
  if data["NOx(GT)"][i]==data["NOx(GT)"].min():
     data.loc[i, 'NOx(GT)'] = data["NOx(GT)"].mean()
for i in range(0,data.shape[0]):
  if data["PT08.S3(NOx)"][i]==data["PT08.S3(NOx)"].min():
     data.loc[i, 'PT08.S3(NOx)'] = data["PT08.S3(NOx)"].mean()
for i in range(0,data.shape[0]):
  if data["NO2(GT)"][i]==data["NO2(GT)"].min():
     data.loc[i, 'NO2(GT)'] = data["NO2(GT)"].mean()
for i in range(0,data.shape[0]):
  if data["PT08.S4(NO2)"][i]==data["PT08.S4(NO2)"].min():
     data.loc[i, 'PT08.S4(NO2)'] = data["PT08.S4(NO2)"].mean()

for i in range(0,data.shape[0]):
  if data["AH"][i]==data["AH"].min():
     data.loc[i, 'AH'] = data["AH"].mean()
for i in range(0,data.shape[0]):
  if data["RH"][i]==data["RH"].min():
     data.loc[i, 'RH'] = data["RH"].mean()       
```

```python

for i in range(0,data.shape[0]):
  if data["T"][i]==data["T"].min():
     data.loc[i,'T'] = data["T"].mean()
for i in range(0,data.shape[0]):
  if data["PT08.S5(O3)"][i]==data["PT08.S5(O3)"].min():
     data.loc[i, 'PT08.S5(O3)'] = data["PT08.S5(O3)"].mean()
for i in range(0,data.shape[0]):
  if data["CO(GT)"][i]==data["CO(GT)"].min():
     data.loc[i, 'CO(GT)'] = data["CO(GT)"].mean()
```

0 for slightly polluted, 1 for moderately polluted, 2 for highly polluted

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3,random_state = 50)
km.fit(data)
```

```python
predict = km.predict(data)
```

```python
#color = np.array(['darkprey','lightsalmon','powderblue'])
plt.scatter(x=data['CO(GT)'],y=data['NMHC(GT)'],c = km.labels_,s=5)
```
